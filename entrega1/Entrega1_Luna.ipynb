### Enunciado
Halle las trayectorias, las líneas de corriente las trazas de una partícula ubicada en $(x_0, y_0)$ a $t = 0$, para los siguientes campos de velocidades bidimensionales:

**(i)** Una corriente uniforme constante superpuesta a otra corriente uniforme ortogonal a la primera. La
velocidad $U'$ de la segunda corriente está modulada en forma armónica en el tiempo con período $\tau$.

**(ii)** Un flujo con campo de velocidad dado por: 

$ \vec{u}(x,y,t) = \frac{\alpha x}{1 + \beta t} \hat{x} + c \hat{y} $

**(iii)** Una fuente lineal de caudal constante superpuesta a una corriente uniforme cuya velocidad U aumenta linealmente con el tiempo.

**(iv)** Un sumidero lineal superpuesto a un torbellino cuya circulación decrece exponencialmente con un tiempo característico $\tau$ .


### Resolución
### (i) 
Tenemos un campo de velocidades bidimensional dado por

$$\vec{u}(x,y) = U_x \hat{x} + U_y \hat{y}$$

Donde $U_y = U_y(t) = U^{\prime} cos(\frac{2 \pi t}{\tau})$

La idea es calcular la trayectoria, traza y líneas de corriente para este campo de velocidad.


#### Trayectoria
Empezando con la trayectoria, integramos en el tiempo las componentes de la velocidad


$$
\begin{aligned}
& \left\{\begin{array}{l}
u_x (t) = \dot{x} = \frac{dx}{dt} = U_x \Rightarrow \int^{x}_{x_0} dx^{\prime}= \int^{t}_{0} U_x dt^{\prime} \\
u_y (t) = \dot{y} = \frac{dy}{dt} = U^{\prime} sin(\frac{2 \pi t}{\tau}) \Rightarrow \int^{y}_{y_0} dy^{\prime} = \int^{t}_{0} U^{\prime} cos(\frac{2 \pi t}{\tau}) dt^{\prime}
\end{array}\right. \\
&
\end{aligned}
$$

Donde quedan las trayectorias en función del tiempo:


$$
\begin{aligned}
& \left\{\begin{array}{l}
x(t)=x_0+u_x t-u_x t \\
y(t)=y_0+\frac{2 \pi U^{\prime}}{\tau}\sin\left(\frac{2 \pi t}{\tau}\right) 
\end{array}\right. \\
&
\end{aligned}
$$
Finalmente, utilizando que $t = \frac{x - x_0}{U_x}$, escribimosla componente de la trayectoria $y$ en función de $x$:

$$
y(x) = y_0+\frac{2 \pi U^{\prime}}{\tau}\sin\left(\frac{2 \pi}{\tau} \frac{(x-x_0)}{U_x} \right)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def tray(x,x0,y0,Ux,Uy,tau):
    return y0 + 2*np.pi*Uy/ tau * np.sin(2*np.pi*(x-x0)/(tau * Ux))

x = np.arange(0,10, 0.1)

plt.plot(x,tray(x,1,1,1,1,1))
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y(x)$', fontsize=14)
plt.title(f'Trayectoria', fontsize=16)
```

#### Corriente

Calculamos las líneas de corriente haciendo uso de la identidad
$$
d\bar{l} \propto \bar{u}=u_x \hat{x}+u^{\prime} \cos (2 \pi t / \varphi) \hat{\varphi} 
$$
Así planteamos la igualdad 
$$ 
\frac{d y}{d x}=\frac{U^{\prime}}{U_x} \cos (2 \pi t / \zeta)
$$
integrando sobre $(x_0 , y_0)$ obtenemos
$$
y(x)=y_0+\int^{x}_{x_0} \frac{U^{\prime}}{U_x} \cos (2 \pi t / \tau) dx^{\prime} \\
\Longrightarrow y(x) = y_0 + \frac{U^{\prime}}{U_x} cos(2\pi t/ \tau)(x - x_0)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def corriente(x,x0,y0,Ux,Uy,t,tau):
    return y0 + Uy/ Ux * (x-x0) * np.cos(2*np.pi*t/tau)

x = np.arange(0,2, 0.1)
for t_ in np.arange(0,10,1):
    plt.plot(x,corriente(x,1,1,1,t_,25,1),label=f't= {np.round(t_,2)}')
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y(x)$', fontsize=14)
plt.title(f'Lineas de Corriente', fontsize=16)
plt.legend()
```

#### Traza $\bar{x}_{\left(t, \bar{x}_0\right)}$
Para las lıneas de traza consideremos que tenemos un punto de referencia fijo $(x_r, y_r)$ en el instante
$t$. Calculamos para un elemento de fluido que en $t = 0$ se encontraba en $(x_0, y0)$ y en el instante $t^{\prime}$ en el punto de referencia $(x_r, y_r)$. Entonces:

$$
\begin{aligned}
& \left\{\begin{array}{l}
x_R=x\left(t^{\prime}\right)=x_0+U_x t^{\prime} \Rightarrow x_0=x_R-u, t^{\prime}\\
y_R=y\left(t^{\prime}\right)=y_0+U^{\prime} \sin \left(\frac{2 \pi t^{\prime}}{\tau}\right) \frac{2 \pi}{tau} \Rightarrow y_0=y_R-U^{\prime} \sin \left(\frac{2 \pi t^{\prime}}{\tau}\right) \frac{2 \pi}{\tau}
\end{array}\right. \\
&
\end{aligned}
$$

Para obtener una expresion $y = y(x)$ despejamos $t^{\prime}$ de las igualdades anteriores y reemplazamos

$$
x(t)=\left(x_R-U_x t^{\prime}\right)+U_x t \quad \Rightarrow t^{\prime}=\frac{x_R-x}{u x}+t
$$
Finalmente llegamos a la expresion
$$
 y\left(x, x_R, t\right)=y_R-U \sin \left(\frac{2 \pi}{\tau}\left[\frac{x_R-x}{U_x}+t\right]\right) \frac{2 \pi}{\tau}+U^{\prime} \sin \left(\frac{2 \pi t}{\tau}\right) \frac{2 \pi}{\tau} 
$$

### (ii)
El campo de velocidades viene dado por:
$$
\bar{u}(x, y, t)=\frac{\alpha x}{1+\beta t} \hat{x}+c \hat{y}
$$

#### Trayectoria:
integramos en el tiempo:
$$
\left\{\begin{array}{l}
u_x=\dot{x}=\frac{d x}{d t}=\frac{\alpha x}{1+\beta t} \Rightarrow \int_{x_0}^x \frac{d x^{\prime}}{x^{\prime}}=\int_0^t \frac{\alpha}{1+\beta t^{\prime}} d t \\
u_y=\dot{y}=\frac{d y}{d t}=c \Rightarrow \int_{y_0}^y d y^{\prime}=\int_0^t c d t^{\prime}
\end{array}\right.
$$
Las trayectorias en función del tiempo quedan:
$$
\left\{\begin{array}{l}
x(t) = x_0 (1+\beta t )^{\alpha/\beta} \\
y(t) = y_0 + ct
\end{array}\right.
$$


escribimos $y$ en función de $x$ :
$$
y(x)=y_0+\frac{c}{\beta}\left[\left(\frac{x}{x_0}\right)^{\beta / \alpha}-1\right]
$$

#### Lineas de corriente

Utilizamos la identidad:
$$
d\bar{l} \propto \bar{u}=\frac{\alpha x}{1+\beta t} \hat{x}+c \hat{y}
$$
Asi planteanos la igualdad
$$
\frac{d y}{d x}=\frac{c(1+\beta t)}{\alpha x} \Rightarrow \int_{y_0}^y d y^{\prime}=\int_{x_0}^x \frac{c(1+\beta t)}{\alpha x^{\prime}} d x^{\prime}
$$
Resolviendo la integral nos queda que
$$
y(x, t)=y_0+\frac{c}{\alpha}(1+\beta t) \ln \left(\left|\frac{x}{x_0}\right|\right)
$$

#### Traza:
$$
\begin{aligned}
& \left\{\begin{array}{l}
x_R=x\left(t^{\prime}\right)=x_0 (1 + \beta t^{\prime})^{\alpha/beta} \Rightarrow x_0=x_R (1 + \beta t^{\prime})^{\beta/\alpha}\\
y_R=y\left(t^{\prime}\right)=y_0+c t^{\prime} \Rightarrow y_0=y_R- c t^{\prime}
\end{array}\right. \\
&
\end{aligned}
$$

Queda
$$
\begin{aligned}
& \left\{\begin{array}{l}
x(t)=x_R(1+\beta t)^{\beta / \alpha}(1+\beta t)^{\alpha / \beta}=x_R\left(\frac{1+\beta t}{1+\beta t^{\prime}}\right)^{\alpha / \beta} \\
y(t)=y_R+\left(1 t-t^{\prime}\right)
\end{array}\right. \\
\end{aligned}
$$
Para obtener $y=y(x)$ despejando $\tau^{\prime}$  de las igualaldes anteriores y reemplazamos:
$$
\Rightarrow y=y_r+\frac{c}{\beta}(1+\beta t)\left[1-\left(\frac{x_1}{x}\right)^{\beta / \alpha}\right]
$$

#### (iv)

Un sumidero lineal superpuesto a un torbellino cuya circulación decrece exponencialmente con un
tiempo característico τ .

$$
\bar{u}(r,\theta) = \frac{-Q}{2 \pi r}\hat{r} + \frac{\Gamma }{2 \pi r} e^{-t/ \tau} \hat{\theta}  
$$

```python
#aca definimos el esquema

def u_5(X,t):
    #llegan coordenadas cartesianas y las convierto a polares
    r, th = cart2polar(X[0],X[1])
    tau = 1
    Q = 1
    gamma = 1
    #funcion de velocidad en coordenadas cartesianas
    u_r = -Q/(2*np.pi*r)
    u_th = gamma/(2*np.pi*r) *np.e**(-t/tau)
    
    #cambio de base de coordenadas polares a cartesianas
    u_x = u_r*np.cos(th) - u_th*np.sin(th)
    u_y = u_r*np.sin(th) + u_th*np.cos(th)
    return np.array([u_x,u_y])

#parametros iniciales y de pasos    
N = 1000
dt = 1e-3
r0 = 0.6
th0s = np.linspace(0,2*np.pi,20)

p0s = np.array([p0_(r0,th0) for th0 in th0s])
#p0s = np.array([[-1, a] for a in np.linspace(-1, 1, 9)])
```

```python
#utilidade

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


def p0_(r,th):
    x = r*np.cos(th)
    y = r*np.sin(th)
    return [x,y]


def polar_to_cart(r,th):
    x = r*np.cos(th)
    y = r*np.sin(th)
    return x,y

def cart2polar(x,y):
    r = np.sqrt(x**2 + y**2)
    if x>0:
        th = np.arctan(y/x)
    elif x<0:
        th = np.arctan(y/x) + np.pi
    elif x == 0 and y > 0:
        th = np.pi/2
    elif x==0 and y<0:
        th = -np.pi/2
    return r,th

def trayectoria(X0, u, N, dt):
    """
    Calcula numéricamente la solución a dX/ds = u(X,t) cómo
    un arreglo, usando el esquema definido previamente

    Parámetros:
    X0: Condición inicial cómo arreglo de dimensión d
    u : Campo de velocidades cómo función de (x,t)
    N : Número de pasos temporales
    dt: Paso temporal
    """
    d = X0.shape[0]            # Dimensión del problema
    ts = np.arange(0, N+1)*dt  # Tiempos donde calcularé la solución
    Xs = np.zeros((N+1, d))    # Solución aproximada
    Xs[0] = X0                 # Impongo la condición inicial a t=0
    for j in range(N):
    # Aplico esquema anterior sobre Xs[j] para obtener Xs[j+1]
        #Xs[j+1] = esquema_midpoint(Xs[j], u, ts[j], dt)
        Xs[j+1] = paso_RK(Xs[j], u, ts[j], dt)
        
    return ts, Xs

def linea_de_corriente(l0, u, t, N, ds):
    """
    Calcula numéricamente la solución a dX/ds = u(X,t) cómo
    un arreglo, usando el esquema definido previamente

    Parámetros:
    l0: Origen de la línea cómo arreglo de dimensión d
    u : Campo de velocidades cómo función de (x,t)
    t : Tiempo para el cual se está calculando la línea
    N : Número de pasos en s
    ds: Paso del parametrizador
    """
    d = l0.shape[0]            # Dimensión del problema
    s = np.arange(0, N+1)*ds   # Valores del parametrizador s
    ls = np.zeros((N+1, d))    # Solución aproximada a lo largo de s
    ls[0] = l0                 # Impongo la condición inicial a s=0
    for j in range(N):
    # Aplico esquema sobre ls[j] para obtener ls[j+1]
        #ls[j+1] = esquema_midpoint(ls[j], u, t, ds)
        ls[j+1] = paso_RK(ls[j], u, t, ds)
    return s, ls


def esquema_midpoint(X, u, t, dt):
  """
  Calcula X(t+dt) a partir de X(t) y el campo de velocidades.

  Parámetros:
    X : Solución a tiempo t (X(t)) cómo arreglo de dimensión d
    u : Campo de velocidades cómo función de (x,t)
    t : Tiempo donde se conoce t
    dt: Paso temporal
  """
  return X + (dt/2)*(u(X,t)+u(X+dt*u(X,t),t+dt)) #Runge Kutta a orden 2

def paso_RK(X, u, t, dt):
    """
    Calcula X(t+dt) a partir de X(t) y el campo de velocidades.

    Parámetros:
    X : Solución a tiempo t (X(t)) cómo arreglo de dimensión d
    u : Campo de velocidades cómo función de (x,t)
    t : Tiempo donde se conoce t
    dt: Paso temporal
    """
    k1 = u(X,t)
    k2 = u(X+k1*dt/2,t)
    k3 = u(X+k2*dt/2,t)
    k4 = u(X+k3*dt,t)
    X = X + (k1 + 2*(k2+k3) + k4)/6*dt
    #t = t + dt
    return X #Runge Kutta a orden 4

def evolucionar_lineas(l0s, ts, u, N, ds, ext=None, figsize=None):
    n = l0s.shape[0]   # Cantidad de líneas de corriente
    m = ts.shape[0]    # Cantidad de tiempos (fotogramas)
    ls = np.zeros((m, n, N+1, 2))
    for i in range(m):
        for j in range(n):
            _, ls[i,j] = linea_de_corriente(l0s[j], u, ts[i], N, ds)
    xs = ls[:,:,:,0]
    ys = ls[:,:,:,1]

    if ext is None:
        extent = [np.min(np.array(xs)), np.max(np.array(xs)),
                  np.min(np.array(ys)), np.max(np.array(ys))]
    else:
        extent = ext

    # Guardo el estado de plt
    params_viejos = plt.rcParams
    plt.rc('animation', html='jshtml')

    fig, ax = plt.subplots(1, 1, figsize=figsize, constrained_layout=True)
    plt.close();  # Cerrar la figura, animation va a crear la suya propia

    # Inicializo las curvas
    plots = [ ax.plot([], [], "b-")[0] for i in range(n) ]
    dots  = [ ax.plot([], [], "ko")[0] for i in range(n) ]
    ax.set_title("$t=0$")
    ax.set_xlabel("x")
    ax.set_ylabel("y")

    def init_():
    
        sup = 0
        for i, (x, y) in enumerate(zip(xs[0], ys[0])):
            dots[i].set_xdata([x[0]])
            dots[i].set_ydata([y[0]])
            sup = max(sup, max(np.max(np.abs(x)), np.max(np.abs(y))))

        ax.set_aspect('equal')
        ax.set_xlim(extent[0], extent[1])
        ax.set_ylim(extent[2], extent[3])

        return plots

    def actualizar_(t):
    
        print(f"\rCalculando fotograma {t} de {m}",
              end="")

        for i, (x, y) in enumerate(zip(xs[t], ys[t])):
            plots[i].set_xdata(x)
            plots[i].set_ydata(y)

        ax.set_title(f"$t={ts[t]:.5f}$")

        return plots

    anim = animation.FuncAnimation(fig, actualizar_, init_func=init_,
                                     frames=range(0, m), blit=True, repeat=True)

        # Restauro el estado de plt
    plt.rc(params_viejos)

    return anim


def evolucionar_trazas(X0s, u, N, dt, paso=1, ext=None, figsize=None):
    n = X0s.shape[0]   # Cantidad de líneas de traza
    ts = np.arange(N+1)*dt
    Xs = np.zeros((N+1, n, N+1, 2))
    for j in range(n):
        Xs[:, j] = X0s[None, j, None]
        for i in range(N):
            for k in range(i//paso+1):
                Xs[i+1, j, k] = paso_RK(Xs[i, j, k], u, ts[i], dt)
    xs = Xs[:,:,:,0]
    ys = Xs[:,:,:,1]

    if ext is None:
        extent = [np.min(np.array(xs)), np.max(np.array(xs)),
                  np.min(np.array(ys)), np.max(np.array(ys))]
    else:
        extent = ext

    # Guardo el estado de plt
    params_viejos = plt.rcParams
    plt.rc('animation', html='jshtml')

    fig, ax = plt.subplots(1, 1, figsize=figsize, constrained_layout=True)
    plt.close();  # Cerrar la figura, animation va a crear la suya propia

    # Inicializo las curvas
    plots = [ ax.plot([], [], "b-")[0] for i in range(n) ]
    dots  = [ ax.plot([], [], "ko")[0] for i in range(n) ]
    ax.set_title("$t=0$")
    ax.set_xlabel("x")
    ax.set_ylabel("y")

    def init():
    
        sup = 0
        for i, (x, y) in enumerate(zip(xs[0], ys[0])):
            dots[i].set_xdata([x[0]])
            dots[i].set_ydata([y[0]])
            sup = max(sup, max(np.max(np.abs(x)), np.max(np.abs(y))))

        ax.set_aspect('equal')
        ax.set_xlim(extent[0], extent[1])
        ax.set_ylim(extent[2], extent[3])

        return plots

    def actualizar(t):
    
        print(f"\rCalculando fotograma {t//paso} de {(N+1)//paso}", end="")

        for i, (x, y) in enumerate(zip(xs[t], ys[t])):
            plots[i].set_xdata(x)
            plots[i].set_ydata(y)

        ax.set_title(f"$t={ts[t]:.5f}$")

        return plots

    anim = animation.FuncAnimation(fig, actualizar, init_func=init,
                                 frames=range(0, N+1, paso),
                                 blit=True, repeat=True)

    # Restauro el estado de plt
    plt.rc(params_viejos)

    return anim

```
